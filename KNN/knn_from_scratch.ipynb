##### Main Logic

```python
def knn(data, query, distance_fn, choice_fn):
    print(query)
    neighbor_distance_and_indices = []
    for index, example in enumerate(data):
        distance = distance_fn(example[:-1], query)
        neighbor_distance_and_indices.append((distance, index))

    sorted_neighbor_distance_and_indices = sorted(neighbor_distance_and_indices)

    k_neighbor_distance_and_indices = sorted_neighbor_distance_and_indices[:k]

    k_nearest_labels = [data[i][-1] for distance, i in k_neighbor_distance_and_indices]
    print(sorted_neighbor_distance_and_indices)

    return choice_fn(k_nearest_labels)
```

```python
def euclidean_distance(point1, point2, p = 2):
    sum_squared_distance = 0
    for i in range(len(point1)):
        sum_squared_distance += math.pow(point1[i] - point2[i], p) 
    return math.sqrt(sum_squared_distance)
     
def mean(labels):
    return sum(labels)/len(labels)

def mode(labels):
    return Counter(labels).most_common(1)[0][0]
```
